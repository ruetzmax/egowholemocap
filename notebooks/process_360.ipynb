```python
import cv2 as cv
import os
```

```python
cap = cv.VideoCapture("vids/down_frontback.mp4")
if not cap.isOpened():
    print("Cannot open vid")
    exit()
    
# clear directory
for filename in os.listdir("processed"):
    file_path = os.path.join("processed", filename)
    if os.path.isfile(file_path) or os.path.islink(file_path):
        os.unlink(file_path)
    
frame_number = 0
while True:
    ret, frame = cap.read()
 
    if not ret:
        break
    
    h, w = frame.shape[:2]
    left_half = frame[:, :w//2]
    right_half = frame[:, w//2:]
    
    cv.imwrite(f"processed/{str(frame_number)}.jpg", left_half)
    
    frame_number+=1
 
cap.release()

print(f"Exported {frame_number} frames with shape ({h}, {w//2})")
```
